```python
import pandas as pd
import json

# Read the JSON file into a Python object
with open('/content/drive/MyDrive/MOF/471mofs_030618.json', 'r') as f:
    data = json.load(f)

# Convert the JSON object to a Pandas DataFrame
df = pd.json_normalize(data)
df_transposed = df.transpose()
# Write the DataFrame to an Excel file
df_transposed.to_csv('data.csv', index=False)
```

```python
C3H6O_rows = df_transposed[df_transposed.index.str.contains('C3H6O_acetone')]
C3H6O_rows.head(5)
```

```python
C3H6O_pr=C3H6O_rows[C3H6O_rows.index.str.contains('pressure')]
C3H6O_ad=C3H6O_rows[C3H6O_rows.index.str.contains('loading')]
```

```python
C3H6O_pr.columns = ['P']
C3H6O_pr.head(5)
```

```python
C3H6O_pr1=C3H6O_pr['P'].astype(str).str.replace('[', '').str.replace(']', '')

C3H6O_pr2 = C3H6O_pr1.to_frame()
C3H6O_pr2.head(5)
```

```python
C3H6O_pr2[['p1','p2','p3','p4']] = C3H6O_pr1.str.split(',', expand=True)
```

```python
C3H6O_pr3 = C3H6O_pr2.drop('P', axis=1)
C3H6O_pr3
```

**Summary of pressure part**

```python
# C2H4 pressure 
C3H6O_pr.columns = ['P']
C3H6O_pr1=C3H6O_pr['P'].astype(str).str.replace('[', '').str.replace(']', '')
C3H6O_pr2=C3H6O_pr1.to_frame()
C3H6O_pr2[['p1','p2','p3','p4']] = C3H6O_pr1.str.split(',', expand=True)
C3H6O_pr3 = C3H6O_pr2.drop('P', axis=1)
C3H6O_pr3
```

**Ad**

```python
C3H6O_ad.columns = ['Ad']
C3H6O_ad1=C3H6O_ad['Ad'].astype(str).str.replace('[', '').str.replace(']', '')
C3H6O_ad2=C3H6O_ad1.to_frame()
C3H6O_ad2[['ad1','ad2','ad3','ad4']] = C3H6O_ad1.str.split(',', expand=True)
C3H6O_ad3 = C3H6O_ad2.drop('Ad', axis=1)
C3H6O_ad3
```

```python
#combine
l1=C3H6O_ad3.values.tolist()
l2=C3H6O_pr3.values.tolist()
for i in range(len(l1)):
    l1[i].extend(l2[i])

C3H6O_new=pd.DataFrame(l1,columns=C3H6O_ad3.columns.tolist()+C3H6O_pr3.columns.tolist())
C3H6O_new.head(5)
```

```python
# change column sequence
new_order = ['p1','ad1', 'p2', 'ad2','p2','p3','ad3','p4','ad3']
C3H6O_update = C3H6O_new.reindex(columns=new_order)

C3H6O_update

#C3H6O_update.to_csv('C3H6O.csv', index=False)
```

```python
C3H6O_in=C3H6O_pr.index
C3H6O_in2=C3H6O_in.str.replace(' ', '').str.replace('Pa', '').str.replace('C2H6', '').str.replace('(', '').str.replace(')', '').str.replace('.', '').str.replace('C3H6O_ethenepressure', '')
print(C3H6O_in2)
```

```python
C3H6O_update.set_index(C3H6O_in2, inplace=True)
C3H6O_update
```

```python
C3H6O_update.to_csv('C3H6O.csv', index=True)
```
